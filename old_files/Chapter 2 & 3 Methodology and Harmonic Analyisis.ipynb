```python
#utility method to track tables and figures in document
class TableFigureCounter:
    def __init__(self):
        self.tab_count = 0
        self.fig_count = 0
    def get_next_tab(self, chapter=None):
        self.tab_count = self.tab_count + 1
        return 'Table ' + str(chapter) + '.' + str(self.tab_count)        
    def get_next_fig(self, chapter=None):
        self.fig_count = self.fig_count + 1
        return 'Figure ' + str(chapter) + '.' + str(self.fig_count)
    def reset(self):
        self.tab_count = 0
        self.fig_count = 0
```

<h4>2.2.3 Tranforming the musicXML data into a searchable database</h4>
<p>MusicXML (a subset of XML) is a a data format data format designed to be both machine readable and human readable and has been growing in popularity since its inception in 2006. All information that would typically be found on a music score can be encoded in MusicXML and software packages such as Sibelius, Logic Pro, and Finale use this export and import sheet music in this format. MusicXML has some similiarities to the MIDI format, however rather than capturing basic information such as tempo, frequency, velocity and rendering this in a binary file, it can store up to 600 attributes that might be found on a score (lyrics, chord symbols, expression etc.) and is human readable (i.e. not stored as 1's and 0's</p>
<p>The solos were exported from MuseScore and Sibelius in the musicXML data format were then processed using the the python programming library called music21 which has been developed at MIT for music analysis. The library includes functionality to parse the musicXML file and convert it into a list of frequencies and the related times at which theyese frequencies occurred. The music21 libary, although having an extensive toolset for music analysis was only used parse the the musicXML data, and to aid in rendering music score examples in the thesis</p>
<p>The Python programming language has a rich set of tools for data analysis and scientific computing which have been utilised for data preparation and analysis. Using the Pandas libary, each note was transformed into record/row in a DataFrame (visually similiar to a row in a Excel spreadsheet). Each row also a range of information about this note (such as which notes preceded and successed it, the current tempo, the current underlying chord etc.)</p>
<p>A sample of the resulting data set can be seen in Table 2.1 where each row of data is a note from one of the improvised solos</p>



```python
mu_stella = mu.converter.parse('xml_files/stella_by_starlight.xml')
mu_autumn = mu.converter.parse('xml_files/autumn_leaves_v2.xml')
mu_if_i_were = mu.converter.parse('xml_files/if_i_were_a_bell.xml')
mu_groovin = mu.converter.parse('xml_files/groovin_high.xml')
mu_all_the_things = mu.converter.parse('xml_files/all_the_things_you_are.xml')
def prep_data_from_xml(xml_file=''):
    notes = mu.converter.parse(xml_file)
    i = 0
    parts_list = []
    while i < len(notes.parts):
        parts_list.append(notes.parts[i])
        i = i + 1


    notes_list = []
    part_name_list = []
    for part in parts_list:
        a = part.flat.getElementsByClass(mu.note.Note)
        b = part.id
        notes_list.append(a)
        part_name_list.append(b)



#create lists
    mel_notes = []
    mel_pos = []

    solo_notes = []
    solo_pos = []

    root_notes = []
    root_pos = []


    third_pos = []
    third_notes = []

    fifth_notes = []
    fifth_pos = []

    seventh_notes = []
    seventh_pos = []

#add notes from parts into lists
    for note in notes_list[0]:
        mel_pos.append(note.offset)
        mel_notes.append(note.midi)
    
    for note in notes_list[1]:
        solo_pos.append(note.offset)
        solo_notes.append(note.midi)

    for note in notes_list[5]:
        root_pos.append(note.offset)
        root_notes.append(note.midi)
    
    for note in notes_list[4]:
        third_pos.append(note.offset)
        third_notes.append(note.midi)
    
    for note in notes_list[3]:
        fifth_pos.append(note.offset)
        fifth_notes.append(note.midi)
    
    for note in notes_list[2]:
        seventh_pos.append(note.offset)
        seventh_notes.append(note.midi)



# <p>Create Data Frames for each part showing location and frequency</p>



    solo_df = pd.DataFrame({'part':0,
                   'location': solo_pos,
                   'note': solo_notes,
                   })

    root_df = pd.DataFrame({'part':1,
                  'location':root_pos,
                  'note': root_notes,
                  })
    third_df = pd.DataFrame({'part':2,
                  'location':third_pos,
                  'note': third_notes,
                  })

    fifth_df = pd.DataFrame({'part': 3,
                         'location':fifth_pos,
                         'note':fifth_notes,
                         })
    seventh_df = pd.DataFrame({'part': 4,
                           'location':seventh_pos,
                           'note':seventh_notes,
                           })



    solo_root_df = solo_df.append(root_df)
    solo_root_third_df = solo_root_df.append(third_df)
    solo_root_third_fifth_df = solo_root_third_df.append(fifth_df)
    solo_root_third_fifth_seventh_df = solo_root_third_fifth_df.append(seventh_df)
    all_note_events = solo_root_third_fifth_seventh_df
    all_note_events = all_note_events.sort('location')


    def create_root_col(note_event):
        if note_event['part'] == 1:
            return note_event['note']
        else:
            return None

    def create_third_col(note_event):
        if note_event['part'] == 2:
            return note_event['note']
        else:
            return None
    
    def create_fifth_col(note_event):
        if note_event['part'] == 3:
            return note_event['note']
        else:
            return None
    
    def create_seventh_col(note_event):
        if note_event['part'] == 4:
            return note_event['note']
        else:
            return None
    
    

    all_note_events['root'] = all_note_events.apply(create_root_col, axis=1)
    all_note_events['third'] = all_note_events.apply(create_third_col, axis=1)
    all_note_events['fifth'] = all_note_events.apply(create_fifth_col, axis=1)
    all_note_events['seventh'] = all_note_events.apply(create_seventh_col, axis=1)


    all_note_events = all_note_events.sort(['location', 'root'], ascending=[True, False])
    all_note_events['root'] = all_note_events['root'].fillna(method="pad")
    
    all_note_events = all_note_events.sort(['location', 'third'], ascending=[True, False])
    all_note_events['third'] = all_note_events['third'].fillna(method='pad')
    
    all_note_events = all_note_events.sort(['location', 'fifth'], ascending=[True, False])
    all_note_events['fifth'] = all_note_events['fifth'].fillna(method='pad')

    all_note_events = all_note_events.sort(['location', 'seventh'], ascending=[True, False])
    all_note_events['seventh'] = all_note_events['seventh'].fillna(method='pad')
    all_note_events['measure'] = (all_note_events['location'] + 4) / 4
    
    solo_note_events = all_note_events[all_note_events.part == 0]

    
    solo_note_events['prep'] = solo_note_events['note'].shift(1)
    solo_note_events['prep'] = solo_note_events['prep'] - solo_note_events['note']
    solo_note_events['res'] = solo_note_events['note'].shift(-1)
    solo_note_events['res'] = solo_note_events['res'] - solo_note_events['note']
    solo_note_events['p_offset'] = solo_note_events['location'].shift(1)
    solo_note_events['p_offset'] = abs(solo_note_events['p_offset'] - solo_note_events['location'])
    solo_note_events['r_offset'] = solo_note_events['location'].shift(-1)
    solo_note_events['r_offset'] = abs(solo_note_events['r_offset'] - solo_note_events['location'])
    
    def find_chord_type(row):
        if row['third'] - row['root'] == 4 and row['fifth'] - row['third'] == 3 and row['seventh'] - row['fifth'] == 3:
            return 'dominant_seventh'
        elif row['third'] - row['root'] == 4 and row['fifth'] - row['third'] == 3 and row['seventh'] - row['fifth'] == 4:
            return 'major_seventh'
        elif row['third'] - row['root'] == 3 and row['fifth'] - row['third'] == 4 and row['seventh'] - row['fifth'] == 3:
            return 'minor_seventh'
        elif row['third'] - row['root'] == 3 and row['fifth'] - row['third'] == 3 and row['seventh'] - row['fifth'] == 4:
            return 'minor_seventh_flat_five'
        else:
            return 'not_determined'

    solo_note_events['chord_type'] = solo_note_events.apply(find_chord_type, axis=1)

    def calc_dist_note_root(data_row):
        x = (data_row['note'] - data_row['root']) % 12
        return x
    
    
    solo_note_events['dist_note_to_root'] = solo_note_events.apply(calc_dist_note_root, axis=1)

    
    
    
    
    return solo_note_events
```

<p>Each row in the resulting data set has 16 attributes. With the exception of the 'name' and 'current tempo' attributes, all of these attributes are derived from the frequency and location information from the musicXML rendering of the music score. A description of the data attributes can be seen below</p>
<ul><b>1. Location:</b>
    <p>Location is a measure of time and describes how many quarternotes any particular note is from the beginning a particular solo</p></ul>
       
<ul><b>2. Note:</b>
    <p>The midi frequency of a each note in the solo.</p></ul>
        
<ul><b>3. Part:</b>
    <p>The score originally contained 5 parts (one for the solo, one for the fifth, seventh, third and root. Although it is not used in the analysis it has been kept in the data set for future work where analysis is across a number of parts</p></ul>

<ul><b>4. Root:</b>
    <p>The midi frequency of the root note of chord over which the solo note is being played</p></ul>
        
<ul><b>5. Third:</b>
        <p>The midi frequency of the third chord over which the solo note is being played</p></ul>    

<ul><b>6. Fifth:</b>
        <p>The midi frequency of the fifth of the chord over which the solo note is being played</p></ul>

<ul><b>7. Seventh:</b>
        <p>The midi frequency of the seventh of chord over which the solo note is being played</p></ul>

<ul><b>8. Measure:</b>
        <p>The measure (and exact location within that measure) of any given note in a solo</p></ul>

<ul><b>9. Prep:</b>
        <p>The distance between midi frequency of a note in a row of data, and the note in the previous row of data</p></ul>

<ul><b>10. Res:</b>
        <p>The distance between midi frequency of a note in a row of data, and the note in the next row of data</p></ul>

<ul><b>11. P_offset:</b>
        <p>This is a measure of time, being distance in quarter notes between a note in a row or data, and the note in the previous row of data. It is used to ascertain whether the previous note is part of same phrase as the currrent note</p></ul>

<ul><b>12. R_offset:</b>
        <p>This is a measure of time, being distance in quarter notes between a note in a row or data, and the note in a note in the row of data. It is used to ascertain whether the next note is part of same phrase as the currrent note</p></ul>

<ul><b>13. Chord Type:</b>
        <p>Calculated by examining the distance between the frequencies of the root, third, fifth and seventh and assigning an appropriately named chord</p></ul>

<ul><b>14. Dist_note_to_root:</b>
        <p>The distance between the midi frequency of the note of the solo and the midi frequency of the root. This is an absolute value and calculated modulo 12</p></ul>

<ul><b>15. Current Tempo</b>
        <p>An approximation, calculated from the recording</p></ul>

<ul><b>16. Name:</b>
        <p>The name of the song from which the solo is taken</p></ul>

<h4>2.3.4 The position on the beat where notes are played is an indicator of note choice</h4>
<p>The placement of a note in relation to the beat is critical to this analysis. This not only relates to the note being designated as on or off the beat, but exactly where on the beat it is occurring (i.e. being between two beat subdivisions in some manner). The overall tempo can be shown to be critical indicator of note choice.</p>


<p>This trend is not only limited to the dataset. Jazz Standards[f] are full of these kinds of idiomatic chord progessions. This allows for direct comparison between many different songs and solos and it becomes possible to examine what note choices different players use on the same porgressions, or which note choices are used by the same player on different progessions. For examples, there are 4 instances of II-V-I progressions in the first 8 bars of the jazz standard, Stella By Starlight, and there are also 5 instances of this progression in the first 16 bars of all the things you are.
